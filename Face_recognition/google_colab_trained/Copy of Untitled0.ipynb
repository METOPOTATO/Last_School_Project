```
import tensorflow as tf
import keras
import numpy as np 
import cv2
from sklearn.model_selection import train_test_split
from keras.applications import VGG16,VGG19
from keras.layers import Dense, Flatten,Dropout
from keras.models import Model,Sequential
from keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers
import pickle
```

```
def save_pickle(obj, file_path):
  with open(file_path, 'wb') as f:
    pickle.dump(obj,f)

def load_pickle(file_path):
  with open(file_path,'rb') as f:
    obj = pickle.load(f)
  return obj
```

```
path_train = path + 'Data/train'
X_train = []
y_train = []
for file in os.listdir(path_train):
  X = load_pickle('pkl_224/data_train_'+ file +'.pkl')
  y = load_pickle('pkl_224/labels_train_'+ file +'.pkl')
  X_train.extend(X)
  y_train.extend(y)
```

```
X_val = load_pickle('pkl_224/data_val.pkl')
y_val = load_pickle('pkl_224/labels_val.pkl')
```

```
for x in X_train:
  x = cv2.dnn.blobFromImage(x, 1.0, IMG_SIZE, [104, 117, 123], False, False)
  mean,std =  x.mean(), x.std()
  x = (x - mean) / std
  
for x in X_val:
  x = cv2.dnn.blobFromImage(x, 1.0, IMG_SIZE, [104, 117, 123], False, False)
  mean,std =  x.mean(), x.std()
  x = (x - mean) / std
```

```
from sklearn.preprocessing import LabelEncoder
encode = LabelEncoder()
encode.fit(y_train)
y_train = encode.transform(y_train)
y_val = encode.transform(y_val)
```

```
X_train = np.asarray(X_train)
y_train = np.asarray(y_train)
X_val = np.asarray(X_val)
y_val = np.asarray(y_val)
```

```
VGG_model = VGG19(
      include_top = False,
      weights='imagenet',
      input_shape = IMG_SHAPE
  )
model = Sequential()

for layer in VGG_model.layers:
  layer.trainable = False
  model.add(layer)

model.add(Flatten())
model.add(Dense(1024, activation='relu') )
model.add(Dense(1024, activation='relu') )
# model.add(Dropout(0.2))
model.add(Dense(512, activation='relu') )
model.add(Dense(7, activation='softmax') )


model.summary()
```

```
epochs = 10
checkpoint_path = 'checkpoints/'
callbacks = [
    keras.callbacks.ModelCheckpoint(
      filepath=checkpoint_path,
      save_freq='epoch',
      save_best_only=True,
      verbose=1
    ),
]
model.compile(
    optimizer=keras.optimizers.Adam(1e-3),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"],
)

model.fit(
    X_train,
    y_train,
    epochs=epochs,
    validation_data=(X_val,y_val),
    callbacks=callbacks
)
```
